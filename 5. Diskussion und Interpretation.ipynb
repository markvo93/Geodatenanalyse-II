# 1.1 Univariate Methoden

Die univariate Analyse ist ein erster wichtiger Schritt in der jeder Datenanalyse. Es wird eine unabhängige Messgröße analysiert und verschiedene Parameter berechnet. Dazu gehören Lage- und Streuungsparameter. Für meine Analyse habe ich mir die Parameter der Grundwassertemperatur und den Sauerstoffgehalt verwendet. 

### Lagemaße beschreiben die zentrale Tendenz der Daten und bestehen aus Arithmetisches Mittel, Median und Quartile

#### 1) Das arithmetische Mittel ist das gebräuchlichste Mass der zentralen Tendenz 

Bei der GWT liegt das a.Mittel bei 12.56°C

Bei dem O2-Gehalt liegt das a.Mittel bei 2.34 mg/l

#### 2) Der Median teilt eine Stichprobe in zwei gleich grosse Hälften. Er ist das 50%-Quantil der Verteilung einer Variablen

Bei der GWT liegt der Median bei 12.3°C

Bei dem O2-Gehalt liegt der Median bei 1.5 mg/l

Im Vergleich zum arithmetischen Mittel ist der Median nicht von Extremwerten(Ausreißern) beeinflusst, deswegen unterscheidet sich hier auch der Wert beim O2-Gehalt so stark 

#### 3) Die Quartile sind Werte, die eine Stichprobe von Daten in vier gleiche Teile teilen

Die Quartile bei der GWT: 11.4, 12.3, 13.4

Die Quartile bei O2: 0.6, 1.5, 3.3 

### Die Streuungsparameter beschreiben die Variabilität der Ausprägung eines Merkmals in einem Datensatz und bestehen aus Range, Varianz, Standardabweichung und IQR

#### 1) Die Range beschreibt die Differenz aus dem Maximum - Minimum Wert 

Bei der GWT liegt die Range bei 17.4°C

Bei dem O2-Gehalt liegt die Range bei 11.3 mg/l

Zu erkennen ist, dass die Range deutlich über dem Mittelwert liegt.
Die Range verliert aber je nach größe des Datensatzes an Informationsgehalt und ist nicht robust gegen Ausreißern. 

#### 2) Der Interquartilbereich (IQR) ist der Abstand zwischen dem ersten Quartil (Q1) und dem dritten Quartil (Q3)

Bei der GWT liegt der IQR bei 2. Das bedeutet, dass 50 % aller Punkte zwischen 11.4°C und 13.4°C liegen. Aber aufgrund der hohen Spannweite der Daten lässt sich darauf schließen, dass die Streubreite der Daten relativ hoch ist. 

Bei dem O2-Gehalt liegt der IQR bei 2.7. Das bedeutet, dass 50 % aller Punkte zwischen 0.6 und 3.3 mg/l liegen. Die Interpretation gestaltet sich wie bei der Temperatur. Eine hohe Range der Daten führt zu einer größeren Streubreite. 

#### 3) Die Varianz wird als durchschnittliche quadratische Abweichung der einzelnen Beobachtungswerte vom arithmetischen Mittel errechnet

Bei der GWT liegt die Varianz bei 2.56332

Bei dem O2-Gehalt liegt die Varianz bei 5.72599

Ohne die Standardabweichung ist die Varianz schwer zu interpretieren, da es durch die Quadrierung eine andere Einheit besitzt, als der ursprüngliche Wert. Dazu muss die Standardabweichung hinzugezogen werden. Allerdings je größer die Varianz, desto größer ist die Streubreite der Daten.

#### 4) Die Standardabweichung ist die Quadratwurzel der Varianz und Maß für die Streuung bzw. die Streubreite der Daten um den Mittelwert

Bei der GWT liegt die Standardabweichung bei 1.60104	

Bei dem O2-Gehalt liegt die Standardabweichung bei 2.3929

Ein höherer Wert bei der Standardabweichung verweist auf eine größere Streubreite der Daten und beim Vergleich zwischen Temperatur und Sauerstoffgehalt, besitzt letzterer Wert eine größere Streubreite der Daten.

### Charakterisierung der Stichproben durch Schiefe und Wölbung

#### 1) Die Schiefe gibt das Ausmaß an, in dem die Daten asymmetrisch sind.

Bei der GWT liegt die Schiefe nach Pearson bei 0.162506

Bei dem O2-Gehalt liegt die Schiefe nach Pearson bei 0.352093

Beide Werte liegen > 0, d.h die Daten für beide Parameter sind rechtsschief. 

#### 2) Wölbung

Bei der GWT liegt die Wölbung bei 3.8061

Bei dem O2-Gehalt liegt die Wölbung bei 2.14896

Beide Werte liegen > 0, d.h die Daten für beide Parameter sind steilgipflig.

# 1.3 Multivariate Statistik

#### Analyse und Visualisierung von mehrdimensionalen Daten

Im Zuge der multivariten Statistik wurde die Clusteranalyse kMeans durchgeführt, denn bei der Hauptkomponentenanalyse besteht das Problem, dass die PCA keine Klassen im Datensatz besitzen. Deshalb kann man die Hauptkomponenten mit keiner Klasse in Beziehung setzen und das Ergebnis ist falsch. 

Bei der Clusteranalyse musste darauf geachtet werden, dass man nicht das komplette Datenset benutzt, sondern die Parameteranzahl verringert, da sonst, durch zu viele Werte fehlen. 

#### Diskussion der Ergbnisse

1) Die Clusteranalyse von O2-Gehalt und Grundwassertemperatur hat zwei Cluster ergeben. Es zeigt sich, dass bei einem O2-Gehalt um die 10 mg/l, sich die Temperatur zwischen 10.5 und 11.5°C bewegt. Das zweite Cluster dagegen beschreibt eine höhre Streuung der Temperatur bei niedrigen O2-Gehalten. 

2)Gleichermaßen Ergbnisse hat die Clusteranalyse von pH-Wert und Grundwassertemperatur ergeben. Hier zeigt sich, ein Cluster bei niedriger Temperatur und pH-Werten zwischen 6.2 und 6.8 und das zweite Cluster weist eine höhe Streuung auf. 

3) Im Vergleich dazu liefert die Clusteranalyse von PO4 und NH4 keine brauchbaren Ergebnisse. Die Cluster sind gehen ineinander über. Der Ausreißer könnte der Grund dafür sein (bei ca. 3mg/l PO4). 

Als mögliche Fehler kommt die Verwendung von einer ungeeigneten Anzahl von Cluster als Startwerte infrage, dort können sich unter Umständen komplett andere Lösungen oder sogar ungeeignete Clustereinteilungen ergeben. Auch problematisch für die Clusteranalyse sind Datenmengen, die sich überlappen oder in Teilen ineinander übergehen. In diesen Fällen ist das k-Means-Verfahren nicht in der Lage, die verschiedenen Gruppen zuverlässig voneinander zu trennen, wie man bei der dritten Clusteranalyse sieht. Als Lösung könnte man die verwendeten Daten noch verkleinern und auf ein paar Parameter reduzieren.

# 3. Interpolationsmethoden

### Determinitische Interpolationsmethoden

##### 1. Lineare Interpolation

lineareIntp.png



Zu beobachten ist eine klarer Anstieg der Pegelstände Richtung Süden und südlich von Karlsruhe. Hingegen scheinen die Pegelstände in nordwestlichter Richtung abzunehmen. 
Zudem scheint die lineare Interpolation den räumlichen Bereich nordöstlich von Karlsruhe, in dem keine Messstellen vorhanden sind, sinnvoll interpoliert worden zu sein. Obwohl man dazu sagen muss, dass es in der Realität natürlich anders aussehen könnte. 

Auswahl der linearen Interpolation, weil die Interpolation dort die räumlichen Unterschiede am besten zeigt und ein klarer Trend zu erkennen ist. 

##### 2. Radiale Basisfunktion Interpolation

Rbf.png

Aus Interesse wurde im Vergleich dazu eine RBF Interpolation erstellt, die nicht zu den gemessen Pegelständen passt. Da gibt es keine klare Interpolation an den gemessen Punkten. Der Grund dafür könnte die Größe der Skala sein, da die Temperaturen bis 160 °C dargestellt sind, was aber an der Interpolationsmethode liegt. So zeigt die Interpolation ausßerhalb der Messstellen eine höhere GWT an, die nicht logisch erscheint. Die lineare Interpolationsmethode liefert da schon bessere Ergebnisse. 

RBF_begrenzt.png

Auch die Eingrenzung der Skala bewirkt keinen Effekt zur Verbesserung der Interpolation. Die Außenbereiche des interpolierten Gebietes steigt an, wobei innerhalb keine sichtbare Interpolation zu sehen ist. Ein möglicher Grund könnte ein zu kleines Raster sein. 

### Kriging Interpolationsmethoden

#### 1. Ordinary Kriging

Ordinary Kriging.png

Ich habe Ordinary Kriging ausgewählt, weil hier meist von einem vorrangingen Trend in den Daten ausgegangen wird und außerdem die am häufigsten verwendete Kriging Methode ist. 
Weil bei Grundwassertemperaturen von Trends in deren Verteilung auszugehen ist und es mögliche wissenschaftliche Erkärungen dafür gibt, habe ich mich für Ordinary Kriging entschieden.
Als mögliche Erklärungen zu Unterschieden beim Grundwasserpegel könnte die Topografie, die Abfolge der Grundwasserleiter und Nichtleiter und die damit zusammenhängende Grundwasserstockwerksgliederung und andere Gründe, dienen. Auch die jeweilige Nutzung und Förderung des Grundwassers spielt eine Rolle. Aber anhand der gemessenen Grundwasserpegel und der Interpolation erscheint das Ergebnis passend.

Wie auch bei der linearen Interpolation ist ein klarer Trend von steigenden Pegeln, von Nord nach Süd, zu erkennen. Allerdings ist der räumliche Bereich ohne Messstellen anders interpoliert und auch hier ist es ohne Werte schwierig zu sagen, ob die Interpolation gut oder schlecht ist. 

#### Interpretation:
Betrachtet man die Interpolationsergebnisse der Grundwasserpegel so kann festgestellt werden, dass die lineare Interpolation die besten Ergebnisse liefert. Das Verfahren zeigt die geringsten Abweichungen zu den gemessenen Pegelständen. Das Kriging Verfahren nähert sich zwar dem linearen Interpolationsverfahren an, lässt aber auch einige Abweichungen erkennen. 

Bei der Radial Base Funktion werden keine verwendbaren Ergebnisse erzielt.
